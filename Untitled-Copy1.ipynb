как мы заметили ранее distance_km и duration_min высокая корреляция - 0.96, соответственно нам не нужны оба этих признака для построения модели, так как они оба содержат одинаковую информацию, удалим признак duration_min из датасета.

```python
data = data.drop(['duration_min'], axis=1)
data.head()
```

```python
data.drop(['driver_latitude', 'driver_longitude', 'origin_order_latitude', 'origin_order_longitude'], axis =1, inplace = True)
```

```python

from sklearn.preprocessing import LabelEncoder, OneHotEncoder

label_enc = LabelEncoder()
oh_enc = OneHotEncoder()
```

```python
ohe_class = oh_enc.fit_transform(data[['offer_class_group']]).toarray()
ohe_class = pd.DataFrame(ohe_class, columns=oh_enc.categories_[0])
ohe_class
```

```python
ohe_ride = oh_enc.fit_transform(data[['ride_type_desc']]).toarray()
ohe_ride = pd.DataFrame(ohe_ride, columns=oh_enc.categories_[0])
ohe_ride
```

```python
data = data.reset_index()
```

```python
data = data.join(ohe_class, how = 'left')
```

```python
data
```

```python
data = data.join(ohe_ride)
```

```python
data.drop(['offer_class_group', 'ride_type_desc'], axis=1, inplace=True)
```

```python
headers = list(data.columns.values)
```

```python
data.head()
```

```python
from sklearn import preprocessing

from sklearn.preprocessing import MinMaxScaler, StandardScaler

min_max_scaler = preprocessing.MinMaxScaler(feature_range=(0,1))

np_scaled = min_max_scaler.fit_transform(data)
data_norm = pd.DataFrame(np_scaled)
data_norm.columns = headers 
data_norm
```

```python
data_norm.info()
```

```python
output_y = data_norm['driver_response']
input_x = data_norm[['weekday_key', 'hour_key', 'distance_driver_passanger_LAT', 'distance_driver_passanger_LONG', 'distance_km', 'Delivery', 'Economy', 'Kids', 'Premium', 'Standard', 'VIP', 'VIP+', 'XL', 'SMB', 'affiliate', 'business', 'private']]
```

```python
from sklearn.model_selection import train_test_split

rand_state = 1000

X_train, X_test, y_train, y_test = train_test_split(input_x, output_y, test_size=0.25, random_state=rand_state)
```

```python
data
```

```python

from sklearn.ensemble import RandomForestClassifier
# выбираем 100 деревьев в качестве параметра
model=RandomForestClassifier(n_estimators=100)
# обучаем модель
model.fit(X_train,y_train)
```

```python
!pip install matplotlib
```

```python
import seaborn as sns
from matplotlib import pyplot as plt
from sklearn.ensemble import RandomForestClassifier

headers = list(X_train.columns.values)

feature_imp = pd.Series(model.feature_importances_,index=headers).sort_values(ascending=False)

f, ax = plt.subplots(figsize=(6, 6))
sns.barplot(x=feature_imp, y=feature_imp.index)

plt.xlabel('Важность атрибутов')
plt.ylabel('Атрибуты')
plt.title("Наиболее важные атрибуты")
plt.legend()
plt.show()
```

```python
feature_imp = pd.Series(model.feature_importances_,index=headers).sort_values(ascending=False)

f, ax = plt.subplots(figsize=(10, 10))
sns.barplot(x=feature_imp, y=feature_imp.index)

plt.xlabel('Важность атрибутов')
plt.ylabel('Атрибуты')
plt.title("Наиболее важные атрибуты")
plt.legend()
plt.show()
```

```python

from sklearn.metrics import classification_report
model_pred = model.predict(X_test)

print(classification_report(y_test, model_pred))

model.score(X_test, y_test)
```
